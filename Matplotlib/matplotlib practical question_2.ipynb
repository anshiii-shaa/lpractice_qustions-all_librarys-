```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Correlation Matrix for the Iris Dataset
• Load the iris dataset.
• Create a heatmap to visualize the correlation between the numerical variables
(sepal_length, sepal_width, petal_length, petal_width).
• Use the coolwarm colormap and add annotations for clarity.

```python
df=sns.load_dataset("iris")
df
num_iris=df[["sepal_length","sepal_width","petal_length","petal_width"]]
sns.heatmap(num_iris.corr(),annot=True,cmap="coolwarm",fmt=".2f")
```

2. Analyzing the Penguins Dataset

* Load the penguins dataset.

* Create a heatmap for the correlation matrix of numerical columns
(bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g).

* Format the annotation to display two decimal places.

```python
penguins=sns.load_dataset('penguins')
features=penguins[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
corr_matrix=features.corr()
sns.heatmap(corr_matrix,annot=True,fmt='0.2f')
plt.show()
```

3. Correlation Heatmap for the Titanic Dataset

* Load the titanic dataset.

* Create a heatmap for the correlation matrix of the numerical variables
(age, fare, sibsp, parch).

* Use the YlGnBu colormap and set the figure size to (10, 8).


```python

titanic=sns.load_dataset('titanic')
features=titanic[['age','fare','sibsp','parch']]
corr_matrix=features.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,fmt='0.2f',cmap='YlGnBu')

plt.show()
```

5. Diamond Dataset Analysis

* Load the diamonds dataset.

* Calculate the correlation matrix of numerical features
(carat, depth, table, price, x, y, z).

* Create a heatmap using the magma colormap and add a title.

```python
diamonds=sns.load_dataset('diamonds')
features=diamonds[['carat','depth','table','price','x','y','z']]
corr_matrix=features.corr()
sns.heatmap(corr_matrix,annot=True,fmt='0.2f',cmap='magma')
plt.title('Diamond Dataset')

plt.show()
```

6. Exploring Tips Dataset

* Load the tips dataset.

* Create a heatmap to visualize the correlation matrix for numerical variables
(total_bill, tip, size).

* Use the viridis colormap with linewidths of 0.5 to separate cells.


```python
tips=sns.load_dataset('tips')
features=tips[['total_bill','tip','size']]
corr_matrix=features.corr()
sns.heatmap(corr_matrix,annot=True,fmt='0.2f',cmap='viridis',linewidth=0.5)

plt.show()
```

7. Pairwise Correlation Heatmap for MPG Dataset

* Load the mpg dataset.


* Calculate the correlation matrix for the numerical variables
(mpg, cylinders, displacement, horsepower, weight, acceleration).

* Create a heatmap using a diverging colormap like Spectral.

```python
mpg=sns.load_dataset('mpg')
features=mpg[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
corr_matrix=features.corr()
sns.heatmap(corr_matrix,annot=True,fmt='0.2f',cmap='Spectral')

plt.show()
```
