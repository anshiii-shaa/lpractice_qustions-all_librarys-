```python
import seaborn as sns
```

# **1,Swarm Plot with Hue Use the penguins dataset to create,Add hue for gender (sex) to visualize how gender affects body mass within species.**

```python
df=sns.load_dataset("penguins")
df

sns.swarmplot(y="body_mass_g",x="species",data=df,hue="sex")
```

# 2,Overlaying Swarm Plot on Boxplot
 **Load the iris dataset and create a boxplot to compare sepal length (sepal_length) across species
(species),Overlay a swarm plot on the boxplot to highlight individual data points..**

```python
df=sns.load_dataset("iris")
df
sns.swarmplot(y="sepal_length",x="species",data=df)
sns.boxplot(y="sepal_length",x="species",data=df)
```

# 3, Categorical Data and Numerical Data
**Using the titanic dataset, create a swarm plot for passenger age (age) by passenger class (class).Add hue="sex" to differentiate between males and females.**

```python
df=sns.load_dataset("titanic")
df
sns.swarmplot(x="class",y="age",hue="sex",data=df)
```

# 4, Combining Swarm Plot with Violin Plot
**Load the diamonds dataset and create a violin plot for diamond price (price) across cut quality (cut).Overlay a swarm plot for detailed visualization of individual points.**

```python
df=sns.load_dataset("diamonds")
df
sns.swarmplot(x="cut",y="price",data=df)
sns.violinplot(x="cut",y="price",data=df)
```

# 5, Adjusting Swarm Plot Appearance
**Using the tips dataset, create a swarm plot for tip amount (tip) by smoker status (smoker).Adjust the size and palette parameters to customize the appearance.**

```python
df=sns.load_dataset("tips")
df
sns.swarmplot(x="smoker",y="tip",size=15,palette="coolwarm",data=df)
```

# 6, Handling Large Datasets
**Load the penguins dataset and create a swarm plot for flipper length (flipper_length_mm) by species.Use the dodge parameter to separate data points clearly when adding hue (hue="island").**

```python
df=sns.load_dataset("penguins")
df
sns.swarmplot(x="species",y="flipper_length_mm",hue="island",dodge=True,data=df)
```
