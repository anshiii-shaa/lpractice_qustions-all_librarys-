Problem 4: Visualizing Train-Test Split
1. Load the Iris dataset using Scikit-learn.
2. Split the dataset into training and testing sets (70% training, 30% testing) using
train_test_split.
3. Visualize the training and testing data distributions:
○ Plot a scatter plot of two features (sepal length vs. sepal width) for both training
and testing sets.
○ Use different colors or markers to distinguish the training and testing data.

```python
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
df
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
data=sns.load_dataset("iris")
print(data)
train_data, test_data = train_test_split(data,test_size=0.3,random_state=42)

sns.scatterplot(train_data["sepal_length"],color="b",label="train_data")
sns.scatterplot(test_data["sepal_length"],color="orange",label="test_data")
plt.legend()
```

```python
sns.scatterplot(train_data["sepal_width"],color="b",label="train_data")
sns.scatterplot(test_data["sepal_width"],color="orange",label="test_data")
plt.legend()
```
