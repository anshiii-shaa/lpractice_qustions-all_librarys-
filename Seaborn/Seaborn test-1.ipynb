```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1)

df=sns.load_dataset("iris")
df
sns.scatterplot(x="petal_length",y="sepal_width",data=df,color="purple")
plt.show()
```

```python
# 2)
df=sns.load_dataset("planets")
df
sns.lineplot(x="year",y="number",data=df,linestyle="--")
plt.show()
```

```python
# 3)
df=sns.load_dataset("iris")
df
sns.countplot(x="species",data=df,hue="species")
plt.show()
```

```python
# 4)
df=sns.load_dataset("exercise")
df
sns.histplot(x="pulse",data=df,hue="diet",multiple="stack",kde=True)
plt.show()
```

```python
# 5)
df=sns.load_dataset("planets")
sns.violinplot(y="distance",hue="method",data=df,split=True,legend=False)
plt.show()
```

```python
# 6)

df=sns.load_dataset("penguins")
df
plt.figure(figsize=(10,7))
sns.boxplot(x="body_mass_g",hue="island",data=df)
plt.title("boxplot")
```

```python
# 7)
df=sns.load_dataset("flights")
df
sns.kdeplot(x="passengers",data=df,fill=True)
sns.rugplot(x="passengers",color="c",height=0.3,data=df)
```

```python
# 8)

df=sns.load_dataset("iris")
df
sns.barplot(y="sepal_length",hue="species",data=df)
```

```python
# 9)
df=sns.load_dataset("iris")
df
sns.pairplot(data=df,diag_kind="hist",kind="kde")
```
