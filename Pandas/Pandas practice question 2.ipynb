19. Write a function to clean a DataFrame by filling missing data, fixing
incorrect types, and removing duplicates.

```python
def clean(df):
  df= df.fillna(df.mean())
  df= df.drop_duplicates()
  df= df.astype(int)
  return df


import pandas as pd
import numpy as np


df = pd.DataFrame({'A': [1, 1, None], 'B': [4, 4, 5], 'C': [4, 4, 5]})


# print(df)
print(clean(df))
```

# **Aggregation Functions:**

21. Create a DataFrame and use the agg() method to calculate the sum
and mean of numeric columns.

```python
import pandas as pd
import numpy as np


data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily',
'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

df = pd.DataFrame(data)
newdf=df.select_dtypes(include='number')
c=newdf.agg(["sum"])
d=newdf.agg(["mean"])
print(c,"\n")
print(d,"\n")
```

22. Create a DataFrame and use the agg() function to apply multiple
aggregation operations (e.g., min, max, mean) on a specific column.

```python
import pandas as pd

data = {'score': [12.5, 9, 16.5,9],
        'attempts': [10,1,2,1]}
df=pd.DataFrame(data)
c=df.agg(["min","max","mean"])

print(c)

```

23. Create a DataFrame and use the agg() method with a custom
